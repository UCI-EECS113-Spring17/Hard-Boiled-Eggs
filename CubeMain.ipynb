```python
from Cube import Cube
from CubeSolver import CubeSolver
from CubeReader import CubeReader
from MovePrinter import MovePrinter
from pynq import Overlay

Overlay("base.bit").download()

def main():
    reader = CubeReader()
    cube = CubeSolver(reader.readCube())
    printer = MovePrinter(Cube())
    # cube._scramble()
    # printer.print_moves(cube.moves)
    cube.moves = []
    cube.solve()
    printer.print_moves(cube.moves)
    
if __name__ == '__main__':
    main()
```
